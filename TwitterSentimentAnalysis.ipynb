```
from tweepy import Stream
from tweepy import StreamListener
import json
import csv
import re
from textblob import TextBlob
```

```
sentiment_level=0

header=['word']
with open('Twittersentiment.csv','w') as file:
  writer=csv.DictWriter(file,header)
  writer.writeheader()


class Listener(StreamListener):
  def on_data(self,data):
    raw_tweets=json.loads(data)
    try:
      tweets=raw_tweets['text']
      tweets = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweets).split())
      tweets = ' '.join(re.sub('RT',' ', tweets).split())
    except:
      print("")
    blob=TextBlob(tweets.strip())
    word_sentiment=0
    for sent in blob.sentences:
      word_sentiment=word_sentiment + sent.sentiment.polarity
    with open('Twittersentiment.csv','a') as file:
      writer=csv.DictWriter(file,header)
      info={'word': word_sentiment}
      writer.writerow(info)

    print(tweets)
    print()


  def on_error(self,status):
    print(status)
```

```
twitter_stream=Stream(Auth,Listener())
twitter_stream.filter(track=['vardy']) 
```
