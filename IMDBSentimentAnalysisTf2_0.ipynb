```
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_datasets as tfdb
from tensorflow.keras import layers
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
tfdb.list_builders()
```

```
data=tfdb.load(name='imdb_reviews')
```

```
(train_data,validate_data),test_data= tfdb.load(name='imdb_reviews',split=(tfdb.Split.TRAIN.subsplit([6,4]),tfdb.Split.TEST),as_supervised=True)
```

```
train_examples,train_lables=next(iter(train_data.batch(10)))
```

```
train_examples
```

```
train_lables
```

```
word2vect=hub.load("https://tfhub.dev/google/tf2-preview/gnews-swivel-20dim/1")
```

```
hub_layer=hub.KerasLayer(word2vect,input_shape=[],dtype=tf.string)
```

```
train_examples[:1]
```

```
model=tf.keras.Sequential()
model.add(hub_layer)
model.add(layers.Dense(16,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
model.summary()
```

```
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```
model.fit(train_data.shuffle(10000).batch(512),epochs=25,validation_data=validate_data.batch(512),verbose=1)
```

```
model.predict(test_data.batch(1))
```
